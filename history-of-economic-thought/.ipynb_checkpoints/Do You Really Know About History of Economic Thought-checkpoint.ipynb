# 一、石头缝里迸出的经济学

　　本文为原创文章，作者王耀，转载请联系作者。  

　　四年前，当高中历史课程上播放纪录片《大国崛起》的时候，我们都很振奋，欣赏电视节目的同时也顺便复习功课。四年后的今天，我却多了很多感慨。在主流教育里，经济学仿佛是从斯密那里陡然出现的学科，像是从石头缝里迸出来一样突然。确实，一般民众会知道哪些斯密以前的西方经济学家？嘿，不许拿万能的亚里士多德说事儿！  

　　毫无疑问，斯密的名字已经和现代经济学创始人的身份紧紧联系在了一起，真相是什么？我希望用这篇简短的文章告诉你，斯密对经济学的发展是否做出了正面的巨大贡献还存在疑问，而没有疑问的是，他的思想几乎没有原创性可言。不过，本文主要目的是要带领读者走进真正的经济学思想史，把握思想分歧的本质，而不是生硬地给学派贴标签。事实上，本文并不按照学派来组织论述。  

# 二、经济思想能吃吗

> 经济学家和政治哲学家们的思想，不论它们在对的时候还是在错的时候，都比一般人所设想的更有力量。的确，世界就是由它们统治着。讲求实际的人自认为他们不受任何学理的影响，可是他们经常是某个已故经济学家的奴隶。——《就业、利息与货币通论》  

　　我对上面凯恩斯的这段话笃信不疑，既得利益之势力，未免被人过分夸大，实在不如思想逐渐侵蚀力之大。那么，** 经济思想往往是在什么地方产生根本性分歧，某个学派一旦成为主流，又是如何塑造世界的呢？**     

## 1、商人不可能没有原罪

　　一切从古希腊开始。当西方人第一次将理性的运用从自然界转到人类本身和社会组织上时，时代的政治架构的偏差和偏见开始起作用，从而产生了第一个重大分歧：对私有财产、自由贸易和商人态度的分歧。柏拉图认为城邦利益高于一切，公有制符合正义，他在《国家篇》中第一次详细叙述了社会中劳动分工的重要性，论证清晰有力，完全不比亚当·斯密逊色；他的弟子亚里士多德虽然承认人是城邦的动物，但却偏向** 私有财产** 制度，认为私有财产符合人的多样性，具有更高的生产力，但是依然反对自由贸易。   

　　对自由贸易和商人的仇视在早期基督教根深蒂固而又矛盾重重，《旧约全书》谴责对于货币的追求，并宣称一个商人不可能没有原罪，但是就在同一本《传道书》中，读者又被教导不要为了利润和商业成功而感到耻辱。最终直到伟大的教父圣奥古斯丁（公元354-430）那里，商人的原罪才被洗清，并对后世产生深刻影响，在12、13世纪被反复引用。  

　　中世纪繁盛时期，意大利的** 博洛尼亚大学** 成为了教会法与罗马法的研究源头和大本营，类似地，巴黎大学成为神学的研究中心。教会法、教令法、罗马法包括查士丁尼的《法典》都只在博洛尼亚大学得以全面发展，并相互影响，进而渗透到西欧其他地方。对商人和贸易的态度在复杂、曲折的演进过程中趋于友好。最终，这个分歧在哲学家、神学家圣托马斯·阿奎那（1225-1274）那里盖棺定论，当代经济学流派也早已经达成了共识：私有财产不受侵犯，商人也是价值创造者。  

　　需要解释一下，虽然马克思主义经济学把服务等非物质生产活动排除在经济核算之外，并创建物质产品平衡表体系(MPS)与西方国民账户体系(SNA)相对，但是由于马克思本人说道“流通是生产的延续”，因此MPS把商业归为五大物质生产部门之一。  

## 2、等价交换

　　罗马帝国时代，经济学思想中第二个重大分歧逐渐露出端倪，并且比第一个分歧复杂得多。罗马法在公元1世纪到3世纪获得了经典性的发展，私法发展了绝对的私有产权理论、自由贸易理论和契约理论，尽管罗马帝国在第4和第5世纪崩溃了，但它的法律遗产却得以延续。在西方具有广泛影响的《狄奥多西法典》和东方大部头的四卷本《民法大全》都致力于自由放任，竭力强调买卖双方自由、自愿地讨价还价所达成的任何价格都是公正的。然而，《民法大全》规定如果卖主以低于其“公平价格”的一半卖出财产时会遭受巨大损失，因而有权找回差价，后来发展成为对自由讨价还价进行攻击的依据。    

　　因此，是否存在这个所谓的** 公平价格** ，或者说是否存在等价交换是经济学史上第二个重大分歧，并广泛影响了人们的思维方式，直到今天分歧实质上仍未解决。亚里士多德应该是最早提出了等价交换理论，并被斯密和马克思引用，而圣托马斯·阿奎那发展了他的分析，除了效用以外又将劳动和费用作为交换价值的一个决定因素，为成本价值论甚至劳动价值论提供了口实。而另一派则最早以圣奥古斯丁教父为代表，他的《神国论》明确提出，** 人们对物品的支付是由于他们的* 需要* 而不是其他任何更为客观的标准或它们的性质的排序决定的** ，这至少是后来主观价值论的基础，而且被包括阿奎那在内的，几乎所有的中世纪神学家、经院学者以及晚期西班牙经院学者所继承。因此，亚当·斯密后来发展出的成熟的等价交换理论，实际上是打破了在欧洲传承了至少1200年的自由定价理论。  

## 3、高利贷禁令

　　首先需要解释清，中世纪及中世纪以前，无论利息多高，凡是放贷取利都称为高利贷。穆瑞·罗斯巴德评论道：除了希伯来人以外，古代世界没有任何一个国家禁止利息。激进的中世纪基督教徒对于高利贷的攻击是令人惊讶的。反对高利贷的运动开始于325年尼西亚的第一届教会会议，会议抓住《旧约全书》里的一段话：  

> 耶和华啊，谁能寄居你的帐篷？他是不放贷取利的人。  

　　实际上它是全书唯一一处反对高利贷的地方。对于所有高利贷的完全禁止第一次被写入世俗的立法中，还是在查理曼皇帝大一统的集权王国中实现的。在789年的帝国宗教会议上，查理曼对王国里的每一个人发布了高利贷禁令。于是，经济思想史中第三个重大分歧浮出水面，高利贷是否有违道德？对于这个看似简单的道德问题进行争论，却引出了分配理论、时间偏好、机会成本等等重要的经济学课题，实际上我认为，一切利益分配上的分歧都附属于这个话题的讨论。争论此起彼伏地持续了一千多年，过程曲折复杂，一言难尽，总之到了最后如罗斯巴德所言：  

> 法学家和神学家们已经完全认识到了收取利息在几乎所有方面都具有正当性：作为一种风险的隐形利润，作为一种投资盈利机会的放弃，以及许多其他情况。但是为什么对于一种简单的、无风险的、不涉及机会放弃的贷款仍然需要收取利息？这个问题直到19世纪后期奥地利学派出现以前，没有人完整给出答案。

　　现实情况是，中世纪以来高利贷的活动从来没有停息过，高利贷禁令逐渐沦为形式。无论是经院学派的黄金时代（13世纪的西欧）还是小阳春时代（16世纪的西班牙），高利贷在形式上依然被顽固地禁止，这部分地导致了经院学派16世纪之后的衰落，人们指责经院学者们为逃避现实的伪君子。罗斯巴德评论道：  

> 为什么在已经有了苏门哈特等人的重要著作支持高利贷的情况下，教会总是倾向于牢固地坚决不改原则，而始终要按照一贯方式行事？** 这是因为在一个不变的外壳形式掩盖下变化其内容，长期以来不仅是天主教会的特征，而且也是任何长寿的官僚机构所具有的特征。**   

　　说到官僚主义传统，中国恐怕更为典型，在不变的形式外壳下更换内容在封建中国还少吗，当代中国呢？这里不予评论，但这种黏性的文化特质值得注意。  

## 4、民族国家

　　关于民族国家的讨论与以往不同，终于可以从15世纪说起，但是又必须涉及绝对主义与世俗主义，以及民族国家发展历程，宗教改革历程这几个话题。简单概括就是：14世纪的大萧条到了15世纪后半页开始让位于经济复苏，西班牙和葡萄牙成为16世纪占据支配地位的民族国家和帝国，而英法则受惠于1453年两国间百年战争的终结，先前很大程度上限于意大利城邦的绝对国家学说，现在扩展到了所有民族国家。到了17世纪，绝对论在欧洲各地最终大获全胜，这个胜利由于16世纪兴起的新教主义和世俗主义而增强了。  

　　14世纪的大萧条可谓深重，1340年到1450年这110年内，西欧人口减少了三分之一，黑死病是罪魁祸首，但国家政府新施加的统治也是重要原因。“百年战争”中，国家通过征税与管制，造成了大灾难，具体参见阎照祥的[《英国史》](http://www.yuntu.io/book/0201917882776#catalogue) 总之在这个时期，民族国家与教会间的平衡被打破，教会权利被摧毁，财产被掠夺，民族国家占据了支配地位。因此，托马斯主义不可避免地衰落了，毕竟自然法要求国家限制自己，使其服从神法或自然法的支配。绝对论出现后，则要求尘世支配神灵，实证法律支配自然法。绝对论的代表人物是马西利乌斯，他著有[《和平保卫者》](http://www.yuntu.io/book/2060907784746#catalogue)   

　　16世纪对经院学派的攻击来自两个派别：新教徒和世俗主义者，前者抨击托马斯主义的理性运用，后者向专制主义疯狂献媚，因此经院学派急剧衰落了。  

　　读者肯定能猜到，关于民族国家和专制主义的论述最终会导向 [重商主义](http://mp.weixin.qq.com/s?__biz=MzI0NDA3MzU3Nw==&mid=2455360857&idx=1&sn=98c287136e455f8ef40f0cc694342439&scene=4#wechat_redirect)，没错，但是我关注的是这个学派的根源及性质，而不是它的观点，他的观点是一文不值的。经济学思想史中第四个分歧其实是个立场分歧，重商主义实质上是为国家代言的一群人聚在一起，天真地相信一国的苦难就是另一国的福利，因而全面规制经济来搞贸易顺差，结果是逼急了法国农民，惹怒了东欧农奴，还差一点把英国经济拖垮，当然葡萄牙和西班牙此时已经被害的落伍了。重商主义另一个作用：** 捧红了亚当·斯密 **，留着下文再说。

## 5、研究方法

　　以上四种分歧具备同一种特点，即在历史上都以伦理分歧的形式出现：商人是否有原罪？自由交换价格不遵循“公平价格”是否算合法？高利贷是否道德？国家权力是否可以质疑？  

　　但是研究方法的分歧不属于观点分歧，却从古至今十分普遍，应该更偏重于方法论上的分歧。当今世界形形色色的经济学流派，主要是在两个方面出现分歧：研究方法分歧和立场分歧。方法论分歧从毕达哥拉斯学派到如今的数理经济学、行为经济学，涉及面很广，一言难尽，暂且不表。

# 三、 亚当·斯密是怎么火起来的

　　罗斯巴德认为，现代经济学的奠基人是理查德·坎替隆（1680s—1734年），而并不是后来的亚当·斯密（1723—1790），这个观点在比较过两人的代表作《商业性质概论》和《国富论》之前存疑，但是比《国富论》早40年写作、早20年出版的《商业性质概论》已经覆盖了现代经济学所有主要理论，提倡自由放任，而且预示了边际革命。  

　　亚当·斯密的分工理论在柏拉图时代就已经萌芽；货币理论不如坎替隆深刻；价值论则打破了欧洲1200多年的传统，成为了李嘉图和马克思的祖师爷，后来被边际主义替代，后者有被消费者行为理论替代，因此斯密的价值论是否是把经济学引向了错误的方向，见仁见智；工资理论和人口理论被罗斯巴德说成“逐字逐句抄袭了坎替隆”；而自由放任的思想在当时比比皆是，并不是斯密独创，那么斯密到底是怎么火起来的？  

　　首先，第一次工业革命正好开始于斯密写作《国富论》的年代，英国迅速崛起成为日不落帝国，很有必要将自由放任旗帜送给一个学术领袖，斯密被选中了。因此，英国的影响力有多大，斯密的影响力就有多大。  

　　其次，重商主义的严重缺陷在18世纪早已经暴露无遗，批判重商主义、引领自由放任的美差让斯密得到了，以至于现代人经常将重商主义结束的日子定在《国富论》出版的那一天，一本书终结一个学派的想法是英雄主义化的愚蠢想法，但是最终事实就是，重商主义有多愚蠢，斯密的理论就被认为有多正确。  

　　最后，斯密身边有一批好友，例如大卫·休谟勋爵，让他经常接触英国上流社会，据说英国首相皮特在一次宴会上为迟到的斯密让座，并自称是斯密的学生。这和坎替隆被克伦威尔驱逐，最后被谋杀的境遇形成鲜明对比。因此，《国富论》一出版，迅速推广，坎替隆的影响跌入低谷。

# 四、后记

　　经济思想史是逐渐演化的过程，很多不出名的小人物都发挥了作用，而非一撮而就。在演化过程中也并不是一直进步，而是可能出现重大倒退，杰出人物的作用往往被夸大，而理论的原创者则默默无闻。因此国内的史学教育擅长分阶段、贴标签、设路标的做法，值得反思。  

　　社会发展得越来越复杂，所有的事物都披上了一层面纱，如果对经济思想史的根本逻辑把握不清，很容易被各种学说搞得晕头转向。例如林毅夫和杨小凯关于[后发优势/劣势](http://mp.weixin.qq.com/s?__biz=MzI0NDA3MzU3Nw==&mid=2455360851&idx=1&sn=554b30935106e833866a8271bc2d32d1&scene=4#wechat_redirect)的争论，无论各自怎么说，明眼人都能看出这是立场分歧。而关于立场分歧很多时候是无解的，需要关注现实需要，也就是弄清什么理论是“不变的外壳形式”即过度发展了的权威，阻碍了“实质内容”的进步，从而适当刹车，俗话说过犹不及，是一样的意思。所以只争论优劣势没什么意思，关键是要解决问题。  

　　笔者学术水平很低，对经济思想史了解并不深入，只是发觉现实中人们对经济史普遍存在以上两点误区，因此写作本文。文中的观点肯定有不少错误和偏颇，希望读者能够尽量批评我的观点。
